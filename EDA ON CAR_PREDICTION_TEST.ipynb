```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare data
X = df1[['Present_Price']]
y = df1['Selling_Price']

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Calculate R-squared
r_squared = model.score(X, y)
print(f"R-squared (coefficient of determination): {r_squared:.4f}")
```
