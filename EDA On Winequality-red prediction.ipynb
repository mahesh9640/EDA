### Ploting Histograms for all features

```python
plt.figure(figsize=(20,15))
for i,column in enumerate(df.columns):
    plt.subplot(4,3,i+1)
    sns.histplot(df[column],kde=True,bins=30,color='skyblue')
    plt.title("Distribution of {column}")
plt.tight_layout()
plt.show()
```

### Correlation Heatmap

```python
plt.figure(figsize=(12,10))
corr=df.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f',square=True)
plt.title("correlation Heatmap")
plt.show()
```

### Boxplots of key features vs quality

```python
features=['alcohol']
plt.figure(figsize=(20,12))
for i, feature in enumerate(features):
    plt.subplot(2,3,i+1)
    sns.boxplot(x='quality',y=feature,data=df,palette="Set3")
    plt.title(f'{feature} vs Quality')
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# List of features to test
features = ['alcohol', 'volatile acidity', 'citric acid', 'residual sugar', 'pH']

# Draw Q-Q plots for each feature
for feature in features:
    plt.figure(figsize=(6, 4))
    stats.probplot(df[feature], dist="norm", plot=plt)
    plt.title(f"Q-Q Plot of {feature}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
