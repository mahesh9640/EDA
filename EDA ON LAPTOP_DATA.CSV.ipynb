### Univariate Analysis

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
    
    print(f"\nStatistics for {col}:")
    print(df[col].describe())
```

### Categorical Features

```python
cat_cols = df.select_dtypes(include=['object', 'category']).columns

for col in cat_cols:
    plt.figure(figsize=(8,4))
    if len(df[col].unique()) > 10:
        # For high cardinality, show top 10
        df[col].value_counts().nlargest(10).plot(kind='bar')
    else:
        df[col].value_counts().plot(kind='bar')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
    
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts())
```

### 4. Bivariate Analysis

### Numerical vs Numerical

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Pairplot for key numerical features (sample 3-4 important ones)
sns.pairplot(df[num_cols[:4]])  # Adjust based on your columns
plt.show()
```

### Categorical vs Numerical

```python
for cat_col in cat_cols[:5]:  # Analyze first 5 categorical columns
    for num_col in num_cols[:3]:  # Against first 3 numerical columns
        plt.figure(figsize=(10,6))
        if len(df[cat_col].unique()) > 5:
            # For many categories, show boxplot of top 5
            top_categories = df[cat_col].value_counts().nlargest(5).index
            filtered_df = df[df[cat_col].isin(top_categories)]
            sns.boxplot(x=cat_col, y=num_col, data=filtered_df)
        else:
            sns.boxplot(x=cat_col, y=num_col, data=df)
        plt.title(f'{num_col} by {cat_col}')
        plt.xticks(rotation=45)
        plt.show()
```

### 5. Multivariate Analysis

```python
# Example: Price distribution by two categorical variables
if 'price' in df.columns and len(cat_cols) >= 2:
    plt.figure(figsize=(12,6))
    sns.barplot(x=cat_cols[0], y='price', hue=cat_cols[1], 
                data=df, estimator='median', ci=None)
    plt.title(f'Price by {cat_cols[0]} and {cat_cols[1]}')
    plt.xticks(rotation=45)
    plt.show()
```
