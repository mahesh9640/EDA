### 2. Batting Analysis

```python
print("Batting Analysis")
batter_runs=df1.groupby(['match_id','batter'])['batsman_runs'].sum().reset_index()
top_score=batter_runs.sort_values('batsman_runs',ascending=False).head(5)
print("Top 5 Run Scorers:")
print(top_score)
print("-----------------------")
print("boundary analysis")
boundaries=df1[df1['batsman_runs'].isin([4,6])]
boundary_counts=boundaries['batsman_runs'].value_counts()
print(boundary_counts)
```

```python
# 3. Bowling Analysis
print("\n===== Bowling Analysis =====")
# Most wickets
wickets = df1[df1['is_wicket'] == 1]
top_bowlers = wickets['bowler'].value_counts().head(5)
print("\nTop 5 Wicket Takers:")
print(top_bowlers)

# Economy rates
bowling_stats = df1.groupby('bowler').agg({
    'ball': 'count',
    'total_runs': 'sum',
    'is_wicket': 'sum'
}).rename(columns={'ball': 'balls_bowled', 'is_wicket': 'wickets'})
bowling_stats['overs'] = bowling_stats['balls_bowled'] / 6
bowling_stats['economy'] = bowling_stats['total_runs'] / bowling_stats['overs']
bowling_stats = bowling_stats[bowling_stats['overs'] >= 1].sort_values('economy')
print("\nBest Economy Rates (min 1 over bowled):")
print(bowling_stats.head(5)[['economy', 'wickets']])
```

```python

# 4. Match Analysis
print("\n===== Match Analysis =====")
match_totals = df1.groupby(['match_id', 'inning', 'batting_team'])['total_runs'].sum().unstack().reset_index()
print("\nTeam Totals by Match:")
print(match_totals)

# 5. Dismissal Analysis
print("\n===== Dismissal Analysis =====")
dismissal_counts = wickets['dismissal_kind'].value_counts()
print("\nDismissal Types:")
print(dismissal_counts)

# 6. Extra Analysis
print("\n===== Extra Analysis =====")
extras = df1[df1['extra_runs'] > 0]
extra_types = extras['extras_type'].value_counts()
print("\nExtra Types:")
print(extra_types)
```

### Visualization 1: Top Run Scorers

```python
plt.figure(figsize=(10,6))
sns.barplot(x='batsman_runs',y='batter',data=top_score.head())
plt.title('Top 5 Runs Scorers')
plt.xlabel("Runs")
plt.ylabel("Batter")
plt.show()
```

### Visualization 2: Dismissal Types

```python
plt.figure(figsize=(10, 6))
dismissal_counts.plot(kind='bar')
plt.title('Dismissal Types Distribution')
plt.xlabel('Dismissal Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Visualization 3: Run Distribution

```python
plt.figure(figsize=(10, 6))
df1['batsman_runs'].value_counts().sort_index().plot(kind='bar')
plt.title('Run Distribution per Delivery')
plt.xlabel('Runs')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 4: Economy vs Wickets
plt.figure(figsize=(10, 6))
sns.scatterplot(x='economy', y='wickets', data=bowling_stats[bowling_stats['overs'] >= 3])
plt.title('Bowling Economy vs Wickets (min 3 overs bowled)')
plt.xlabel('Economy Rate')
plt.ylabel('Wickets')
plt.show()
```

```python
# Visualization 5: Powerplay Analysis (first 6 overs)
powerplay = df1[df1['over'] < 6]
powerplay_runs = powerplay.groupby(['match_id', 'inning'])['total_runs'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.boxplot(x='inning', y='total_runs', data=powerplay_runs)
plt.title('Powerplay Runs Distribution (Overs 1-6)')
plt.xlabel('Inning')
plt.ylabel('Runs')
plt.show() 
```
