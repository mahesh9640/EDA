```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Extract the 'sepal_length' column
sepal_lengths = iris['sepal_length']

# Apply CLT: Take many samples and store the mean of each
sample_means = []

for _ in range(1000):  # Take 1000 samples
    sample = np.random.choice(sepal_lengths, size=30, replace=True)
    sample_means.append(np.mean(sample))

# Plot the distribution of sample means
plt.figure(figsize=(8,5))
plt.hist(sample_means, bins=30, edgecolor='black', color='skyblue')
plt.title("Central Limit Theorem on Sepal Length (Iris Dataset)")
plt.xlabel("Sample Mean of Sepal Length")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Haberman dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/haberman/haberman.data"
df = pd.read_csv(url, header=None, names=["age", "operation_year", "aux_nodes", "survival_status"])

# Sort the age column
x = np.sort(df["age"])

# Calculate CDF values
y = np.arange(1, len(x) + 1) / len(x)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='.', linestyle='none', color='darkblue')
plt.title("CDF of Age (Haberman Dataset)")
plt.xlabel("Age")
plt.ylabel("CDF")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/haberman/haberman.data"
df = pd.read_csv(url, header=None, names=["age", "operation_year", "aux_nodes", "survival_status"])

# Draw the box plot
plt.figure(figsize=(6, 5))
plt.boxplot(df["age"], vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Box Plot of Age (Haberman Dataset)")
plt.ylabel("Age")
plt.grid(True)
plt.show()
```
