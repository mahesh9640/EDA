```python
# Total matches played
print("Total matches:", len(df))

# Seasons covered
print("\nSeasons:", df['season'].unique())

# Teams participated
teams = pd.concat([df['team1'], df['team2']]).unique()
print("\nTeams:", teams)
```

```python
### Mathches Trendas Over Seasons:

plt.figure(figsize=(12,5))
sns.countplot(x='season',data=df,palette='viridis') ### Here viridis means colour
plt.title("Matches Played Per Season") ### Adds a title to the plot.
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['winner'].value_counts().plot(kind='bar')
plt.title("total wins by team")
plt.show()
win_percentage=(df['winner'].value_counts()/pd.concat([df['team1'],df['team2']]).value_counts())*100
win_percentage.sort_values(ascending=False).plot(kind="bar",figsize=(12,6))
plt.title("win percentage by team")
plt.ylabel("win %")
plt.show()
```

### Toss Impact on Match Outcomes

```python
plt.figure(figsize=(8,5))
df['toss_decision'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("toss decision")
plt.show()
```
